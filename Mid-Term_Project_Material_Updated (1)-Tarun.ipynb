## Step 0. Prepare raw input datasets

Here we will 1) load original datasets, 2) remove unused columns, 3) de-duplicate rows, and 4) join datasets, not necessarily in this order

Assumptions: 
- Claim ID (cur_clm_uniq_id) represents one claim, which may or may not have more than one code (code could be HCPCS/CPT, diagnosis, procedure...)
- There is a one-to-many relationship between patient IDs (bene_mbi_id) and claim IDs (cur_clm_uniq_id), i.e. each claim is unique to one patient, but one patient can have more than one claim

### 0.3 Combine all datasets

### 0.3.1. Join datasets with common records

```python
# Join Claims Header and Claims Revenue Center datasets on claim ID, patient ID, and claim date
# Perform outer join to capture all possible claims
medicare_df = pd.merge(
    parta_claims_header_df,
    parta_claims_revenue_center_detail_df,
    on=['claim_id','patient_id','claim_date'], how='outer'
)
medicare_df.sort_values(by='claim_id')#.head(100)
```

```python
# Join Medicare and Diagnosis datasets on claim ID, patient ID, and claim date
# Perform outer join to capture all possible claims
medicare_df = pd.merge(
    medicare_df,
    parta_diagnosis_code_df,
    on=['claim_id','patient_id','claim_date'], how='outer'
)
medicare_df.sort_values(by='claim_id')#.head(100)
```

```python
# Since Claims Header dataset has some principal diagnosis codes and the Diagnosis dataset 
#  supplements them with additional codes wherever possible, 
#  coalesce them with preference to the principal code from Claim Header dataset
# Once the diagnosis codes are combined into one column, remove the older columns and any duplicates
medicare_df['diagnosis_code'] = medicare_df[['prncpl_dgns_cd', 'clm_dgns_cd']].bfill(axis=1).iloc[:, 0]
medicare_df = medicare_df.drop(['prncpl_dgns_cd', 'clm_dgns_cd'], axis=1).drop_duplicates()
medicare_df.sort_values(by='claim_id').head(20)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

cvm_hcpcs_codes = ['36415', '93010', '93000', '93306', '93005', '78452']

# 1. Filter data for CVM claims only
cvm_claims = medicare_df_new[medicare_df_new['hcpcs_code'].isin(cvm_hcpcs_codes)]

# 2. Calculate the total claims and CVM claims per year
total_claims_per_year = medicare_df_new.groupby('claim_year').size()
cvm_claims_per_year = cvm_claims.groupby('claim_year').size()

# 3. Calculate the percentage of CVM claims by year
cvm_share = (cvm_claims_per_year / total_claims_per_year * 100).fillna(0)
print("CVM Claims as a Percentage of Total Claims by Year:\n", cvm_share)

# 4. Calculate the average cost of CVM vs. total claims for each year
avg_cost_total = medicare_df_new.groupby('claim_year')['claim_cost'].mean()
avg_cost_cvm = cvm_claims.groupby('claim_year')['claim_cost'].mean()
cost_comparison = pd.DataFrame({'Total Claims Average Cost': avg_cost_total, 'CVM Claims Average Cost': avg_cost_cvm})
print("\nAverage Claim Cost Comparison by Year:\n", cost_comparison)

# 5. Analyze demographic distribution for CVM claims by age group
bins = [0, 39, 59, 79, 99]
labels = ['<40', '40-59', '60-79', '80+']

if 'patient_age' in medicare_df_new.columns:
    medicare_df_new['age_group'] = pd.cut(medicare_df_new['patient_age'], bins=bins, labels=labels)
    print("Age groups successfully created:")
    print(medicare_df_new[['patient_age', 'age_group']].head())  # Display first few rows to verify
else:
    print("Column 'patient_age' not found in the dataset.")

cvm_age_distribution = medicare_df_new[medicare_df_new['hcpcs_code'].isin(cvm_hcpcs_codes)].groupby('age_group').size()
print("\nCVM Claims Count by Age Group:\n", cvm_age_distribution)


if not cvm_age_distribution.empty:
    plt.figure(figsize=(8, 8))
    cvm_age_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=140, title='CVM Claims by Age Group')
    plt.ylabel('')  # Hide the y-label for a cleaner pie chart
    plt.show()
else:
    print("No data found for CVM claims distribution by age group.")

# 6. Visualize CVM claims as a percentage of total claims by year


remaining_share = 100 - cvm_share  # Non-CVM claims percentage

# Plotting a horizontal stacked bar chart
plt.figure(figsize=(10, 6))
plt.barh(cvm_share.index, cvm_share.values, color='b', label='CVM Claims')  # CVM claims segment
plt.barh(cvm_share.index, remaining_share.values, left=cvm_share.values, color='gray', label='Other Claims')  # Other claims segment

# Add labels and title
plt.title('CVM Claims as a Percentage of Total Claims (2016 - 2018)')
plt.xlabel('Percentage of Claims')
plt.ylabel('Year')
plt.legend(loc='best')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Add text labels for each bar segment
for i, year in enumerate(cvm_share.index):
    # Display CVM claim percentage
    plt.text(cvm_share.values[i] / 2, i, f"{cvm_share.values[i]:.1f}%", 
             ha='center', va='center', color='white', fontweight='bold')
    
    # Display non-CVM claim percentage
    plt.text(cvm_share.values[i] + (remaining_share.values[i] / 2), i, 
             f"{remaining_share.values[i]:.1f}%", ha='center', va='center', color='black', fontweight='bold')

# Show plot
plt.show()


# 7. Visualize average cost comparison for CVM vs total claims
cost_comparison.plot(kind='bar', figsize=(10, 6))
plt.title('Average Claim Cost Comparison (Total vs. CVM) by Year')
plt.xlabel('Year')
plt.ylabel('Average Claim Cost ($)')
plt.xticks(rotation=0)
plt.show()

# 8. Pie chart for CVM claims distribution by age group
plt.figure(figsize=(8, 8))
cvm_age_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=140, title='CVM Claims by Age Group')
plt.ylabel('')  # Hide the y-label for a cleaner pie chart
plt.show()
```
